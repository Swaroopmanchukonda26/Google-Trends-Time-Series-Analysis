```python
# Load the Tesla data
df_tesla = pd.read_csv('TESLA Search Trend vs Price.csv')

# Load the Unemployment data
df_unemployment = pd.read_csv('UE Benefits Search vs UE Rate 2004-19.csv')

# Load the Bitcoin data pieces
df_btc_price = pd.read_csv('Daily Bitcoin Price.csv')
df_btc_search = pd.read_csv('Bitcoin Search Trend.csv')
```

```python
df_btc_price.dropna(inplace=True)
```

```python
df_tesla.MONTH = pd.to_datetime(df_tesla.MONTH)
df_btc_search.MONTH = pd.to_datetime(df_btc_search.MONTH)
df_unemployment.MONTH = pd.to_datetime(df_unemployment.MONTH)
df_btc_price.DATE = pd.to_datetime(df_btc_price.DATE)
```

```python
df_btc_monthly = df_btc_price.resample('M', on='DATE').last()
```

```python
rolling_df = df_unemployment[['UE_BENEFITS_WEB_SEARCH', 'UNRATE']].rolling(window=6).mean()
```

```python
# 1. Set up high-definition size and resolution
plt.figure(figsize=(14, 8), dpi=120)
plt.title('Tesla Web Search vs Price', fontsize=18, pad=15)
plt.xticks(fontsize=14, rotation=45)

# 2. Get our dual axis frames
ax1 = plt.gca()
ax2 = ax1.twinx()

# 3. Add labels and custom presentation colors
ax1.set_ylabel('TSLA Stock Price', color='#E6232E', fontsize=14)
ax2.set_ylabel('Search Trend', color='skyblue', fontsize=14)

# 4. Bind boundaries
ax1.set_ylim([0, 600])
ax1.set_xlim([df_tesla.MONTH.min(), df_tesla.MONTH.max()])

# 5. Plot the lines thickly
ax1.plot(df_tesla.MONTH, df_tesla.TSLA_USD_CLOSE, color='#E6232E', linewidth=3)
ax2.plot(df_tesla.MONTH, df_tesla.TSLA_WEB_SEARCH, color='skyblue', linewidth=3)

plt.show()
```

```python
# 1. Set up canvas size and resolution
plt.figure(figsize=(14, 8), dpi=120)
plt.title('Bitcoin News Search vs Resampled Price', fontsize=18)
plt.xticks(fontsize=14, rotation=45)

# 2. Split axis limits
ax1 = plt.gca()
ax2 = ax1.twinx()

ax1.set_ylabel('BTC Price', color='#F9A602', fontsize=14)
ax2.set_ylabel('Search Trend', color='skyblue', fontsize=14)

ax1.set_ylim([0, 15000])
ax1.set_xlim([df_btc_monthly.index.min(), df_btc_monthly.index.max()])

# 3. Plot Bitcoin price as dashed orange and search trends with circle dots
ax1.plot(df_btc_monthly.index, df_btc_monthly.CLOSE, color='#F9A602', linewidth=2, linestyle='--')
ax2.plot(df_btc_monthly.index, df_btc_search.BTC_NEWS_SEARCH, color='skyblue', linewidth=2, marker='o')

plt.show()
```

```python
# 1. Set up high-definition size and resolution
plt.figure(figsize=(14, 8), dpi=120)
plt.title('Rolling Monthly US "Unemployment Benefits" Web Searches vs UNRATE', fontsize=18)
plt.xticks(fontsize=14, rotation=45)

# 2. Split our dual axes frames
ax1 = plt.gca()
ax2 = ax1.twinx()

# 3. Apply label names and matching custom colors
ax1.set_ylabel('FRED U/E Rate', color='purple', fontsize=16)
ax2.set_ylabel('Search Trend', color='skyblue', fontsize=16)

# 4. Set timeline boundaries and activate the matrix grid
ax1.set_xlim([df_unemployment.MONTH.min(), df_unemployment.MONTH.max()])
ax1.grid(color='grey', linestyle='--')

# 5. Plot actual unemployment as a dashed purple line and the smooth rolling search average
ax1.plot(df_unemployment.MONTH, rolling_df.UNRATE, color='purple', linewidth=3, linestyle='-.')
ax2.plot(df_unemployment.MONTH, rolling_df.UE_BENEFITS_WEB_SEARCH, color='skyblue', linewidth=3)

# 6. Show the complete presentation chart
plt.show()
```
